# Obtaining of sample with exponential distribution by using ICDF

We created a sample, which has Exponencial distribution   
$$p(x) = \lambda^k e^{-\lambda x}$$
$$F(x) = 1-e^{-\lambda x} $$
with $\lambda = 0.5$  
We generated sample $u = \{u_1,u_2,...u_n\}$ from arbitrary distribution U(\[0,1\])  
n = 400000
After that we obtain sample y with Exponencial distribution by using formula:  
$$y = F^{-1}(u)$$
$$y_i = -\frac{1}{\lambda} \ln{(1-u_i)}   \forall i=\overline{1,n}$$
$$y = \{y_1,y_2,...y_n\}$$

# Gibbs sampling

## The obtaining of sapmle with 2-dimentional normal distribution with parameters: correlation $r=0.1,~~m_x=0,~~m_y=0,~~\sigma_x=1,~~\sigma_y=1$

<span style="font-size:20px"> $ f(x|y_i) = f(x,y_i) \\  f(y|x_i) = f(x_i,y)$ </span>

We used accept-reject algoritm, to obrain the one-dimentional samples $X = \{x_1,x_2,...x_n\}$ and $Y = \{y_1,y_2,...y_n\}$  
Size of sample n = 1000  
Initial conditions: $(x_1,y_1)=(1,1)$ 
## One step  
### Firstly, we applied this altoritm to generate $x_2$ with distribution $f(x|y_1)$
For this algoritm the exponential distribution $g(x) = e^{-x}$ was used.
Generation of sample with disribution $g(x)$ was creating by using ICDF algoritm
1. Finding minimum number M, which satisfies the equtaion:  
$Mg(x) - f(x|y_1) > 0$  
By using numerical method - enumerationg of numbers M and $x$. $x \in [0,5]$. For pair normal and exponential distribution with standart parameters this method works correctly.  
2. Obtaining $x_2$ with distribution g(x) by using ICDF  
$$x_2 = -\ln{u_1},~~~u_1 \in U([0,1])$$  
3. Generating $u_2 \in U([0,1])$ and checking condition:  
$u_2 < \frac{f(x_2|y_1)}{Mg(x_2)}$ - $x_2 \in X$    
$u_2 \geq\frac{f(x_2|y_1)}{Mg(x_2)}$ - return to step 2    
4. Generating $u_3 \in U([0,1])$. Checking condition:  
$u_3 < 0.5$ - $x_2 \in X$  
$u_3 \geq 0.5$ - $-x_2 \in X$  
Because all samples from g(x) are positive.  
  
### Secondly, we applied the same algoritm to generate $y_2$ with distribution $f(y|x_2)$.
1. Finding minimum number M, which satisfies the equtaion:  
$Mg(y) - f(y|x_2) > 0~~for~\forall y$   
2. Obtaining $y_2$ with distribution g(y) by using ICDF    
$y_2 = -\ln{u_1},~~~u_1 \in U([0,1])$ 
3. Generating $u_2 \in U([0,1])$ and checking condition:  
$u_2 < \frac{f(y_2|x_2)}{Mg(y_2)}$ - $y_2 \in Y$    
$u_2 \geq\frac{f(y_2|x_2)}{Mg(y_2)}$ - return to step 2    
4. Generating $u_3 \in U([0,1])$. Checking condition:  
$u_3 < 0.5$ - $y_2 \in Y$  
$u_3 \geq 0.5$ - $-y_2 \in Y$  
Because all samples from g(y) are positive.

### After first step we have $X=\{x_1,x_2\}$ and $Y=\{y_1,y_2\}$
### Initial conditions for 2 step is $(x_2,y_2)$. Initial conditions for i step is $(x_i,y_i)$
## We repeat this step (n+100)-times  (sample size n=1000).
## After all steps:
Removing 100 pairs at the beginning $(x_i,y_i),~i=\overline{1,100}$.   
Redesignating of numbering $i \rightarrow i-100$: $\{(x_{101},y_{101}),(x_{102},y_{102}),...,(x_{n+100},y_{n+100})\} \rightarrow \{(x_1,y_1),(x_2,y_2),...,(x_n,y_{n})\} $
