```
# EX2
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-5,5,0.01)
y = x**2 - x - 6
plt.figure(figsize=(25, 5))
sp = plt.subplot(131)
plt.plot(x,y)
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title(r'$y = x^2 - x - 6$',fontsize=16)
plt.grid(True)

x = np.arange(-2.10,-1.90,0.01)
y = x**2 - x - 6
sp = plt.subplot(132)
plt.plot(x,y)
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title(r'$y = x^2 - x - 6$',fontsize=16)
plt.grid(True)

x = np.arange(2.90,3.30,0.01)
y = x**2 - x - 6
sp = plt.subplot(133)
plt.plot(x,y)
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title(r'$y = x^2 - x - 6$',fontsize=16)
plt.grid(True)
# корни: x = -2, x = 3
plt.show()
```

```
#EX3
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-10,10,0.1)
y = np.log((x**2 + 1)*np.exp(-np.abs(x)/10))/np.log(1 + np.tan(1/(1 + np.sin(x)**2)))

plt.plot(x,y)
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.grid(True)
plt.show()
```

```
#EX5
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.05)
y=x**2 + np.random.randn(len(x))*x
plt.figure(figsize=(10, 10))
plt.errorbar(x, y, xerr=0.1, yerr=3, fmt='g. ')
p,sigma = np.polyfit(x,y,deg=1,cov=True)
p_f = np.poly1d(p)
plt.plot(x,p_f(x))
p,sigma = np.polyfit(x,y,deg=2,cov=True)
plt.plot(x,np.polyval(p,x))
plt.legend(loc='best')
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title('График с поргешностями',fontsize=16)
plt.show()
```

```
###EX6
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10,10,0.05)
a = 3
b = 0.4
y = []
for el in x:
  summand = 0
  for n in range(11):
    summand+=b**n+np.cos(a**n*np.pi*el)
  y.append(summand)
plt.figure(figsize=(20,10))
plt.plot(x,y)
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title('График Вейерштрасса',fontsize=16)
plt.show()
```
